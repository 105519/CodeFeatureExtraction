```python
def print_all_tag(repos):
    print(sum([x != [] for x in repos]), '/', len(repos))
    print()
    cnt = dict()
    for x in repos:
        for y in x:
            if (y in cnt):
                cnt[y] += 1
            else:
                cnt[y] = 1
    topics = []
    for x in cnt:
        topics.append((cnt[x], x))
    topics.sort()
    for idx, x in enumerate(topics):
        print(len(topics) - idx, x)

def print_special_tag(repos, special_list):
    cnt = dict()
    for x in repos:
        for y in x:
            if (y in cnt):
                cnt[y] += 1
            else:
                cnt[y] = 1
    topics = []
    for x in cnt:
        topics.append((cnt[x], x))
    topics.sort()
    for idx, x in enumerate(topics):
        if (any(x[1].find(s) != -1 for s in special_list)):
            print(x[0], x[1])

def delete_tag(repos, delete_list):
    new_repos = []
    for x in repos:
        new_repo = []
        for y in x:
            if (y not in delete_list):
                new_repo.append(y)
        new_repos.append(new_repo)
    return new_repos

def change_tag(repos, source, target):
    new_repos = []
    for x in repos:
        new_repo = []
        for y in x:
            if (any(y == s for s in source)):
                new_repo.extend(target)
            else:
                new_repo.append(y)
        new_repos.append(new_repo)
    return new_repos

def clear_tag(repos, limit = 0):
    clear_repos = []
    for x in repos:
        clear_repos.append(list(set(x)))
    cnt = dict()
    for x in clear_repos:
        for y in x:
            if (y in cnt):
                cnt[y] += 1
            else:
                cnt[y] = 1
    new_repos = []
    for x in clear_repos:
        new_repo = []
        for y in x:
            if (cnt[y] >= limit):
                new_repo.append(y)
        new_repos.append(new_repo)
    return new_repos
```
